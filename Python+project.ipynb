## Q7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'and soon....

```python
df['published_month'].replace({1:"January",2: 'February',3:'March',4: 'April',5: 'May',6: 'June',7: 'July',8: 'August',9: 'September',10: 'October',11: 'November',12: 'December'},inplace=True)
df[['published_at','published_month']]
```

## Q8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count..

```python
#Grouping month and sorting descending video count (using pivot table)
df.pivot_table(values='video_id',index='published_month',aggfunc='count').sort_values('video_id',ascending=False)
```

## Q9. Find the count of unique video_id, channel_id and channel_title.

```python
#Getting count of the unique of video_id, channel_id and channel_title.
unique_video_id= df['video_id'].nunique()
unique_channel_id = df['channel_id'].nunique()
unique_channel_title = df['channel_title'].nunique()

#printing the result
print("Total Unique video id is " ,unique_channel_id)
print("Total Unique channel id is ",unique_video_id)
print("Total Unique channel title is", unique_channel_title)
```

## Q10. Find the top10 channel names having the highest number of videos in the dataset and the bottom having lowest number of videos.s.

### Top 10

```python
#Using Pivot table to group month and sorting by video count for Top 10
df.pivot_table(index= 'channel_title', values='video_id', aggfunc='count').sort_values('video_id',ascending= False).head(10)
```

### Bottom 10

```python
#Using Pivot table to group month and sorting by video count for Bottom 10
df.pivot_table(index= 'channel_title', values='video_id', aggfunc='count').sort_values('video_id',ascending= False).tail(10)
```

## Q11. Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.

```python
#using pivot table to find to which video got maximium no of likes.
df.pivot_table(index= 'title', values='likes', aggfunc='max').sort_values('likes',ascending= False).head(1)
```

```python
#using pivot table to find to which video got minimium no of likes.
df.pivot_table(index= 'title', values='likes', aggfunc='max').sort_values('likes',ascending= False).tail(1)
```

## Q13.Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot..

```python
import matplotlib.pyplot as plt
new = df[['view_count','dislikes']]
```

### Correlation

```python
new.corr()
```

### Covariance

```python
new.cov()
```

```python
plt.figure(figsize=(7,5))
plt.ticklabel_format(style='sci')
plt.scatter(new['view_count'],new['dislikes'])
plt.title('View_count v/s Dislikes')
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.style.use(style='ggplot')
plt.show()
```
